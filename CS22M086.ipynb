```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import os
import csv
```

```python
data = pd.read_csv('spam.csv', encoding='latin-1')
data
```

```python
#Removing the last 3 columns which contains no important information
data.drop(columns = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace = True)
```

```python
data.sample(10)
```

```python
# Changing the Label of the Columns
data.rename(columns={'v1' : 'Target' , 'v2' : 'Label'}, inplace = True)
data.sample(10)
```

```python
#Converting the ham / spam string in to 0 / 1 number

encoder = LabelEncoder()
data['Target'] = encoder.fit_transform(data['Target'])

data.sample(5)
```

```python
# using the library nltk for tokenizing and stemming



ham_words = set() 
spam_words = set()
Total_words = set()
count_Spam_words = dict()
count_Ham_words = dict()
ps = PorterStemmer()
for i in range(data.shape[0]):
    text = data.iloc[i]['Label'].lower()
    text = word_tokenize(text)
    # For Spam Email 
    if data.iloc[i]['Target'] == 1:
      for x in set(text):
        if (len(x) > 2) and x.isalnum() : 
          x = ps.stem(x)
          spam_words.add(x)
          if x in count_Spam_words:
            count_Spam_words[x] += 1
          else: 
            count_Spam_words[x] = 1

            count_Ham_words[x] = 1

    else:
    # for Non spam email 
      for x in set(text):
        if (len(x) > 2) and x.isalnum() : 
          x = ps.stem(x)
          ham_words.add(x)
          if x in count_Ham_words:
            count_Ham_words[x] += 1
          else: 
            count_Ham_words[x] = 1

            count_Spam_words[x] = 1

print(spam_words)
print(ham_words)
Total_words = spam_words.union(ham_words)
print(len(Total_words) , len(spam_words), len(ham_words))
# Storing the number of occurances of each of the word for spam and non spam
print(count_Ham_words)
print(count_Spam_words)

# Storing the Total Number of unique words which has occurred in spam and non - spam emails
Total_Spam_Words = len(spam_words)
Total_Ham_Words = len(ham_words)
print(Total_Spam_Words, Total_Ham_Words)
```

```python
for word in count_Spam_words: 
  count_Spam_words[word] = count_Spam_words[word] / Total_Spam_Words

for word in count_Ham_words: 
  count_Ham_words[word] = count_Ham_words[word] / Total_Ham_Words
```

```python
print(count_Ham_words)
print(count_Spam_words)
```

```python
# Method to train the model based on the probabilites of the words given spam or non - spam.
def TestSpamClassifier(email):
  emailWords = email.lower()
  emailWords = word_tokenize(emailWords)
  validWords = set()
  for x in emailWords: 
    if x.isalnum() and len(x) > 2: 
      x = ps.stem(x)
      validWords.add(x)
  # print(validWords)
  result = 1
  # print(Total_words)
  for word in Total_words:
    if word in validWords:
      pham = count_Ham_words[word]
      pspam = count_Spam_words[word]  
      result *= (pspam/pham)

    else :
      pham = 1 - count_Ham_words[word]
      pspam = 1 - count_Spam_words[word]
      result = result *  (pspam/pham)
    # print(pham, pspam)
  result *= (Total_Spam_Words / Total_Ham_Words)
  
  if result > 1:
    # spam email 
    return 1
# Non - spam email
  return 0


result = TestSpamClassifier('Hi, you have won a lottery !! ')

print(result)
```

```python
# Function that reads and appends the data of each text file in the test folder into a list
def read_text_file(file_path):
  global test_mails
  with open(file_path, 'r') as f:
      test_mails.append([f.read()])
```

```python

test_mails = []
# Storing the current path 
curr_path = os.getcwd()
old_path = curr_path
# Migrating to the folder test
curr_path =  curr_path + '/test'
os.chdir(curr_path)

# Reading all of the .txt files in the test folder. 
for file in os.listdir():
    if file.endswith(".txt"):
        file_path = f"{curr_path}/{file}"
        read_text_file(file_path)

# migrating back to the previous folder to store the test_dataset there which store the set of all the email in csv file.
os.chdir(old_path)
filename = "test_dataset.csv"

with open(filename,'w') as csvfile:
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerows(test_mails)
```

```python
# Test for the data
testdf = pd.read_csv("test_dataset.csv",header=None)
testdf = testdf.to_numpy()
test_x = testdf[:,0]
num = testdf.shape[0]
pred_label = []
# Predicting for the set of emails in the test_dataset.csv
for i in range(testdf.shape[0]):
  y_pred = TestSpamClassifier(test_x[i])
  pred_label.append([test_x[i], y_pred])

#Storing result into Test_result.csv
file_name = "test_result.csv"
with open(file_name,'w') as csvfile:
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerow(['Email','Predicted Label'])
    csvwriter.writerows(pred_label)
    
```
